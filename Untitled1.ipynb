```python
import numpy as np

rows1 = int(input("Rows of Matrix A: "))
cols1 = int(input("Columns of Matrix A: "))

A = np.zeros((rows1, cols1), dtype=int)

print("Enter Matrix A elements:")
for i in range(rows1):
    for j in range(cols1):
        A[i][j] = int(input())

rows2 = int(input("Rows of Matrix B: "))
cols2 = int(input("Columns of Matrix B: "))

if cols1 != rows2:
    print("Matrix multiplication not possible!")
else:
    B = np.zeros((rows2, cols2), dtype=int)

    print("Enter Matrix B elements:")
    for i in range(rows2):
        for j in range(cols2):
            B[i][j] = int(input())

    C = A @ B

    print("\nMatrix A:\n", A)
    print("\nMatrix B:\n", B)
    print("\nMatrix Multiplication Result:\n", C).0
```
